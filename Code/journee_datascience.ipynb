```python
# Importer les bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('TableauxdonnéesParcelle.xlsx')
df2 = pd.read_excel('DataVdt.xlsx')
df3 = pd.read_excel("BiomasseMoleculaireMicrobienne.xlsx")
df.head()
```

```python
df2["Nb de taxas"]
```

```python
# Suppression de plusieurs colonnes
colonnes_a_supprimer = df.columns[32:61].tolist() + df.columns[62:65].tolist()
df = df.drop(columns=colonnes_a_supprimer)
df.head()
```

```python
# Remplacer toutes les virgules par des points
df = df.replace(',', '.', regex=True)

#Renommer Parcelle pzr id_parcelle
df = df.rename(columns={'Parcelle': 'id_parcelle'})
df.head()
```

```python
#Nouveau df
df_merged = pd.merge(df, df3[['code echantillon', 'Moy E1_E2']], left_on='id_parcelle', right_on='code echantillon')
df_merged = df_merged.drop(columns="code echantillon")
df_merged.head()
```

```python
df_merged.describe()
```

```python
df_merged.to_excel('/content/TDP.xlsx', index=False)
```

```python
import seaborn as sns

# Select relevant soil properties (for example: Sable, MO, N_Tot, Cu, Zn)
colonnes_sols = df_merged[['Sable', ' MO', 'N_Tot', ' Cu', ' Zn']]

# Create a pairplot to compare all soil properties
sns.pairplot(colonnes_sols)

# Show the pairplot
plt.show()
```

```python
# Créer un scatterplot pour chaque propriété du sol par rapport au nombre d'espèces
plt.figure(figsize=(12, 8))

# Comparaison entre le nombre d'espèces et la teneur en sable
plt.subplot(2, 3, 1)
plt.scatter(df_merged['Species'], df_merged['Sable'], alpha=0.7)
plt.title('Sable vs Nombre d\'espèces')
plt.xlabel('Nombre d\'espèces')
plt.ylabel('Sable')

# Comparaison entre le nombre d'espèces et la teneur en MO
plt.subplot(2, 3, 2)
plt.scatter(df_merged['Species'], df_merged[' MO'], alpha=0.7)
plt.title('MO vs Nombre d\'espèces')
plt.xlabel('Nombre d\'espèces')
plt.ylabel('MO')

# Comparaison entre le nombre d'espèces et la teneur en N_Tot
plt.subplot(2, 3, 3)
plt.scatter(df_merged['Species'], df_merged['N_Tot'], alpha=0.7)
plt.title('N_Tot vs Nombre d\'espèces')
plt.xlabel('Nombre d\'espèces')
plt.ylabel('N_Tot')

# Comparaison entre le nombre d'espèces et la teneur en Cu
plt.subplot(2, 3, 4)
plt.scatter(df_merged['Species'], df_merged[' Cu'], alpha=0.7)
plt.title('Cu vs Nombre d\'espèces')
plt.xlabel('Nombre d\'espèces')
plt.ylabel('Cu')

# Comparaison entre le nombre d'espèces et la teneur en Zn
plt.subplot(2, 3, 5)
plt.scatter(df_merged['Species'], df_merged[' Zn'], alpha=0.7)
plt.title('Zn vs Nombre d\'espèces')
plt.xlabel('Nombre d\'espèces')
plt.ylabel('Zn')

# Ajuster la mise en page
plt.tight_layout()
plt.show()
```
